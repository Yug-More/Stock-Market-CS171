# Classical ML Models for NASDAQ-100  
### Next-Day + Next-Week (5-Day Ahead) Predictions

This notebook is the second part of my CS 171 final project.

In Notebook 1, I created all the features from NASDAQ-100 daily data.  
Now in this notebook, I train several classical machine learning models to predict:

1. **Next-Day Movement** → whether the index closes higher tomorrow  
2. **Next-Week Movement (5 trading days)** → whether close[t+5] > close[t]

The models I use:

- Naive baseline (predict "up" every day)
- Logistic Regression
- Random Forest
- SVM (RBF kernel)

At the end, I make several plots to visualize how well the models perform over time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn models
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# metrics
from sklearn.metrics import accuracy_score, confusion_matrix

np.random.seed(42)

plt.style.use("default")
plt.rcParams["figure.figsize"] = (11, 5)
plt.rcParams["axes.grid"] = True
```

## 1. Load train/val/test datasets

These files were created in Notebook 1 and saved in the `data/` directory.

```python
train_df = pd.read_csv("data/nasdaq100_train.csv")
val_df = pd.read_csv("data/nasdaq100_val.csv")
test_df = pd.read_csv("data/nasdaq100_test.csv")

train_df.head()
```

## 2. Create the 5-day ahead target label

Notebook 1 only included the next-day label (`target_up`).  
So here I create the 5-day ahead version:



I do this separately for the train, val, and test sets.


```python
def make_5day_target(df):
    df = df.copy()
    df["close_5d"] = df["close"].shift(-5)
    df["target_up_5d"] = (df["close_5d"] > df["close"]).astype(int)
    return df.iloc[:-5]  # last 5 rows have no target

train_df_5 = make_5day_target(train_df)
val_df_5 = make_5day_target(val_df)
test_df_5 = make_5day_target(test_df)
```

## 3. Select features + labels for both prediction tasks

```python
feature_cols = [
    "close",
    "ret_1d", "log_ret_1d",
    "ma_5", "ma_10", "ma_20",
    "price_over_ma_5", "price_over_ma_10", "price_over_ma_20",
    "vol_10d", "vol_20d",
    "rsi_14",
]

# Next-day targets
X_train = train_df[feature_cols].values
y_train = train_df["target_up"].values

X_val = val_df[feature_cols].values
y_val = val_df["target_up"].values

X_test = test_df[feature_cols].values
y_test = test_df["target_up"].values

# Next-week targets (5 day horizon)
X_train_5 = train_df_5[feature_cols].values
y_train_5 = train_df_5["target_up_5d"].values

X_val_5 = val_df_5[feature_cols].values
y_val_5 = val_df_5["target_up_5d"].values

X_test_5 = test_df_5[feature_cols].values
y_test_5 = test_df_5["target_up_5d"].values
```

## 4. Scale features (fit on train only)

```python
scaler = StandardScaler()
scaler.fit(X_train)

# scale for next-day models
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# scale for next-week models
X_train_5_scaled = scaler.transform(X_train_5)
X_val_5_scaled = scaler.transform(X_val_5)
X_test_5_scaled = scaler.transform(X_test_5)
```

## 5. Naive Baseline (Predict "Up" Every Day)

```python
y_pred_val_naive = np.ones_like(y_val)
y_pred_test_naive = np.ones_like(y_test)

y_pred_val_naive_5 = np.ones_like(y_val_5)
y_pred_test_naive_5 = np.ones_like(y_test_5)

print("Naive next-day val accuracy:", accuracy_score(y_val, y_pred_val_naive))
print("Naive next-day test accuracy:", accuracy_score(y_test, y_pred_test_naive))
print()
print("Naive 5-day val accuracy:", accuracy_score(y_val_5, y_pred_val_naive_5))
print("Naive 5-day test accuracy:", accuracy_score(y_test_5, y_pred_test_naive_5))
```

## 6. Logistic Regression (Next-Day + Next-Week)

```python
log_reg = LogisticRegression(max_iter=600)
log_reg.fit(X_train_scaled, y_train)

# next-day predictions
val_pred_lr = log_reg.predict(X_val_scaled)
test_pred_lr = log_reg.predict(X_test_scaled)

# next-week predictions
log_reg_5 = LogisticRegression(max_iter=600)
log_reg_5.fit(X_train_5_scaled, y_train_5)

val_pred_lr_5 = log_reg_5.predict(X_val_5_scaled)
test_pred_lr_5 = log_reg_5.predict(X_test_5_scaled)

print("LR next-day test accuracy:", accuracy_score(y_test, test_pred_lr))
print("LR 5-day test accuracy:", accuracy_score(y_test_5, test_pred_lr_5))
```

## 7. Random Forest (Next-Day + Next-Week)

```python
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train_scaled, y_train)

val_pred_rf = rf.predict(X_val_scaled)
test_pred_rf = rf.predict(X_test_scaled)

rf_5 = RandomForestClassifier(n_estimators=200, random_state=42)
rf_5.fit(X_train_5_scaled, y_train_5)

val_pred_rf_5 = rf_5.predict(X_val_5_scaled)
test_pred_rf_5 = rf_5.predict(X_test_5_scaled)

print("RF next-day test accuracy:", accuracy_score(y_test, test_pred_rf))
print("RF 5-day test accuracy:", accuracy_score(y_test_5, test_pred_rf_5))
```

## 8. SVM (Next-Day + Next-Week)

```python
svm = SVC(kernel="rbf", C=1.0)
svm.fit(X_train_scaled, y_train)

val_pred_svm = svm.predict(X_val_scaled)
test_pred_svm = svm.predict(X_test_scaled)

svm_5 = SVC(kernel="rbf", C=1.0)
svm_5.fit(X_train_5_scaled, y_train_5)

val_pred_svm_5 = svm_5.predict(X_val_5_scaled)
test_pred_svm_5 = svm_5.predict(X_test_5_scaled)

print("SVM next-day test accuracy:", accuracy_score(y_test, test_pred_svm))
print("SVM 5-day test accuracy:", accuracy_score(y_test_5, test_pred_svm_5))
```

## 9. Time-Series Prediction Plots  
(Next-Day vs Actual & Next-Week vs Actual)

These are some of the most helpful visualizations for understanding how the model behaves.

```python
plt.figure(figsize=(12,4))
plt.plot(test_df["date"], y_test, label="Actual", alpha=0.7)
plt.plot(test_df["date"], test_pred_rf, label="RF Prediction", alpha=0.7)
plt.title("Next-Day Prediction vs Actual (Random Forest)")
plt.legend()
plt.show()
```

## 10. Rolling Accuracy (Important Visualization)

This shows how accuracy changes over time.

```python
def rolling_acc(actual, pred, window=30):
    return pd.Series(actual == pred).rolling(window).mean()

roll_acc_1d = rolling_acc(y_test, test_pred_rf)
roll_acc_5d = rolling_acc(y_test_5, test_pred_rf_5)

plt.figure(figsize=(12,4))
plt.plot(test_df["date"], roll_acc_1d, label="Next-Day Rolling Accuracy")
plt.plot(test_df_5["date"], roll_acc_5d, label="5-Day Rolling Accuracy")
plt.title("Rolling Accuracy Over Time (Random Forest)")
plt.legend()
plt.show()
```
